```python
from surprise import Dataset
from surprise.model_selection import cross_validate
from surprise import Reader

reader = Reader()
data = Dataset.load_builtin('ml-100k', prompt=False)
```

```python
from surprise import KNNWithMeans, KNNBasic
from surprise.model_selection import GridSearchCV
from surprise.model_selection import train_test_split

# min_support parameter can determine the minimum number of ratings or interactions a user or item needs to have before being considered for recommendations.
sim_options = {
    "name": ["msd", "cosine"],
    "min_support": [1, 5],
    "user_based": [False],
}
param_grid = {
    "k": [10, 25, 40],
    "min_k": [1, 2],
    "sim_options": sim_options
}

gs = GridSearchCV(KNNWithMeans, param_grid, measures=["rmse", "mae"], cv=3)
gs.fit(data)
```

```python
print("GS Best RMSE: ", gs.best_score["rmse"])
print("GS Best MAE: ", gs.best_score["mae"])
```

```python
gs.best_params['rmse']
```

```python
gs.best_params['mae']
```

```python
algo = gs.best_estimator["rmse"]
trainset = data.build_full_trainset()
algo.fit(trainset)
```

```python
testset = trainset.build_anti_testset()

predictions = algo.test(testset)
```

```python
# r_ui = actual rating
# est = estimated rating
for pred in predictions[:25]:
  print(pred)
```
